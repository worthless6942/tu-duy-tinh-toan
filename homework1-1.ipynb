```python
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

pi = 3.14
r = a / 2

dien_tich_chu_nhat = a * b
dien_tich_hinh_tron = pi * r * r

dien_tich_trong_cay = dien_tich_chu_nhat - dien_tich_hinh_tron

print("Diện tích trồng cây:", round(dien_tich_trong_cay, 2))
```

```python
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p-a) * (p-b) * (p-c))
    print("Diện tích:", round(s, 1))
else:
    print("Khong phai 3 canh cua tam giac")
```

bài 8

```python
ten = input("Ten chu ho: ")
truoc = int(input("Chi so thang truoc: "))
sau = int(input("Chi so thang nay: "))

kwh = sau - truoc
tien = 0

if kwh <= 50:
    tien = kwh * 1984
elif kwh <= 100:
    tien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tien = tien * 1.08  # cộng VAT 8%

print("\nHo va ten:", ten)
print("Tien phai tra la:", int(tien))
```
