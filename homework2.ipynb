```python
a = float(input("nhap a: "))
b = float(input("nhap b: "))
c = float(input("nhap c: "))
d = float(input("nhap d: "))
m = float(input("nhap m: "))
n = float(input("nhap n: "))

det = a*d - b*c

if det == 0:
    print("Vo nghiem hoac vo so nghiem")
else:
    x = (m*d - b*n)/det
    y = (a*n - m*c)/det
    print("x =", x)
    print("y =", y)
```

```python
x = float(input("Nhap so giay: "))

so_gio = x // 3600
so_phut = (x - so_gio * 3600) // 60
so_giay = x - so_gio * 3600 - so_phut * 60

print("so gio:", so_gio)
print("so phut:", so_phut)
print("so giay:", so_giay)
```
